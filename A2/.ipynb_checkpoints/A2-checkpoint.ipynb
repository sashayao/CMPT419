```python
#imports
import pandas as pd     #work with data
import matplotlib.pyplot as plt     #plot charts
%matplotlib inline
import seaborn as sns    #make chart pretty
import numpy as np
from sklearn.mixture import GaussianMixture as GMM
import os
sns.set(style="darkgrid")
from math import pi
```

```python
import glob
path = r'intensity'                     
files = glob.glob(os.path.join(path, "*.csv"))    

l = []
# extracting csv files in folder
for f in files:
    df = pd.read_csv(f)
    # labeling videos
    df['video'] = np.full(len(df), int(os.path.basename(f)[:-4]), dtype = int)
    # labeling frames
    df['frame'] = np.arange(1 ,len(df)+1)
    l.append(df)
df = pd.concat(l, ignore_index=True)
df = df.drop(['Unnamed: 0',' face_id'], axis=1)
```

# Part 1
## Cleaning data

```python
# removing all elements with success equal to 0
df = df[df[' success'] != 0]
#removing all elements with confidence less than 0.8
df = df[df[' confidence'] > 0.8]
save = df
```

```python
df.head(5)
```

```python
# extracting a df containing only AUs for data manipulation
Au_data = df.copy()
Au_data = Au_data.drop(['frame',' success', ' confidence', 'video'], axis=1);
```

## Getting BIC

```python
n_components = np.arange(1, 20)
# finding minimum number of componenets
models = [GMM(n, covariance_type='full', random_state=0).fit(Au_data)
for n in n_components]
plt.plot(n_components, [m.bic(Au_data) for m in models], label='BIC')
plt.show()
print("minimum is 11 components")
```

## GMM Clustering

```python
# Au_data = Au_data.drop(['GMM_probs','GMM'],axis=1)
Au_data.head(3)
```

```python
# GMM clustering using 11 components
gmm = GMM(n_components=11, covariance_type='full', random_state=0)
label = gmm.fit(Au_data).predict(Au_data)
# adding cluster grouping frames
df['GMM'] = label

# Addding gmm.prob to frames
probs = gmm.predict_proba(Au_data)
df['GMM_probs'] = probs.max(1)
```

## Finding image closest to Mean

```python
# initiating data to work with
clusters = Au_data.copy()
clusters['GMM_probs'] = probs.max(1)
clusters['GMM'] = label
clusters.head(3)
```

```python
# ordering images according to it's probability of belonging 
# to a cluster, and it's uklidean distance from cluster's Mean
print("Top2 frames closest to mean for 11 clusters")
for i in range(11):
    # put all same data from same cluster into seperate data frames
    cluster = clusters[clusters['GMM'] == i]
    #get mean
    mean = gmm.means_[i]
    cluster = cluster.drop(['GMM', 'GMM_probs'], axis=1)
    # minus difference
    cluster = cluster.sub(mean)
    # using only cluster with probability of one
    cluster['GMM_probs'] = clusters['GMM_probs']
    cluster = cluster.abs()
    # Adding uklidean distance to frames
    cluster["distance"] = cluster.sum(axis=1)
    cluster = cluster.sort_values(['GMM_probs','distance'], ascending=[False,True])
    print(df.loc[cluster.head(2).index][['video','frame']])
```

I have taken the top1-2 frames closest (most representative) of the mean for all clusters. There exist ambiguous clusters. For example, cluster 7 is a poorly classified cluster- where some of the expression involves expressions of outrage, while others are expressions of joy and laughter. I decided to name cluster-7 smiling, since more data the cluster supports the classification of smiling as opposed shock or anger. This occurred because cluster-7 is classified for specific higher values in both Au10 and Au6 (upper-lip raiser and lip-corner puller), Au10 is a signal for expressions disgust and anger while Au6 typically represents smiling, which results in a range of conflicting emotions in the the cluster.

## Radar graph for each cluster

```python
def plotRadar(data, means, weight, clusterName):
    # getting axis and data
    categories=list(data)[0:]
    N = len(categories)
    values=means[:1].flatten().tolist()
    values += values[:1]
    print(values)


    # making radar chart interval 
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    length = len(means)
    a = 70
    # # # Initialise the spider plot
    for i in range(1,length+1):
        plt.figure(figsize=(400/a, 400/a), dpi=a)
        values=means[i-1:i].flatten().tolist()
        values += values[:1]
        ax = plt.subplot(111, polar=True, )

        # # Draw one axe per variable + add labels labels yet
        plt.yticks(size=7)
        plt.xticks(angles[:-1], categories, size=7)

        # # Plot data
        ax.plot(angles, values, linewidth=1, linestyle='solid')
        w = "C-"+str(clusterName[i-1])+" weight "+str(round(weight[i-1],4))
        ax.fill(angles, values, 'g', alpha=0.5)
        plt.title(w, size=11 , y=1)
        plt.show()
```

```python
weight = gmm.weights_
means = gmm.means_
plotRadar(Au_data, means, weight, np.arange(1,12))
```

# Part 2
## UMAP

```python
import umap
from sklearn.preprocessing import StandardScaler
```

```python
# initializing data to work with
data = Au_data.copy()
# normalizing data
reducer = umap.UMAP()
scaled_data = StandardScaler().fit_transform(data)
# reducing data into 2 dimensions
embedding = reducer.fit_transform(scaled_data)
embedding.shape
```

```python
umap_data = df.drop(['GMM','GMM_probs'],axis=1)
umap_data['x'] = embedding[0:,0]
umap_data['y'] = embedding[0:,1]
umap_data.head(3)
```

## Storing data to avoid randomization

```python
# umap_data.to_csv("umap.csv")
umap_data = pd.read_csv('umap.csv')
umap_data = umap_data.drop(['Unnamed: 0'], axis=1)
```

```python
n_components = np.arange(1, 20)
models = [GMM(n, covariance_type='full', random_state=0).fit(umap_data[['x','y']])
for n in n_components]
plt.plot(n_components, [m.bic(umap_data[['x','y']]) for m in models], label='BIC')
plt.show()
```

Best number of component is 13

```python
umap_gmm = GMM(n_components=13, covariance_type='full', random_state=0)
label = umap_gmm.fit(umap_data[['x','y']]).predict(umap_data[['x','y']])
umap_data['GMM'] = label
umap_data['probs'] = umap_gmm.predict_proba(umap_data[['x','y']]).max(1)
```

```python
plt.scatter(
    umap_data['x'],
    umap_data['y'],
    c=label,
    cmap='Oranges'
)

x = umap_gmm.means_[0:13, 0]
y = umap_gmm.means_[0:13, 1]

for i in range(13):
    plt.annotate(str(i+1), (x[i], y[i]))
```

```python
# finding data closest to each cluster's mean
indexes = []

for i in range(13):
    umap_cluster = umap_data[umap_data['GMM'] == i]
    umap_cluster = umap_cluster.sort_values(['probs'], ascending=False)
    indexes.append(umap_cluster.index[0])
```

```python
path = r'test'                     # use your path
files = glob.glob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent

l = []
for f in files:
    test = pd.read_csv(f)
    test['video'] = np.full(len(test), int(os.path.basename(f)[:-4]), dtype = int)
    test['frame'] = np.arange(1 ,len(test)+1)
    l.append(test)

test = pd.concat(l, ignore_index=True)
test = test.drop(['Unnamed: 0',' face_id'], axis=1);
```
